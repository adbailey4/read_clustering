# Determine bandwidth

*Estimate_bandwidth function*

* Takes the input point to be clustered and returns the bandwidth parameter as float. Requires a "quantile" value between 0 and 1 that indicates how the distances between points are handled.

* "Quantile" is given by the np.quantile function

* The function giving the optimal bandwidth value is integrated with the mean_shift method in the VariantClass methods.
